```python
def create_consensus_channel(df):
    """
    Creates a new 'consensus' channel in a pandas DataFrame based on the 
    most frequent label across all columns for each row.

    Args:
        df (pd.DataFrame): The input DataFrame.

    Returns:
        pd.DataFrame: The DataFrame with the new 'consensus' channel added.
    """

    def get_consensus_label(row):
        labels = row.tolist()
        counts = Counter(labels)
        most_common = counts.most_common()

        if not most_common:
            return np.nan

        max_count = most_common[0][1]
        potential_labels = [label for label, count in most_common if count == max_count]

        return potential_labels[0]

    df['yasa_PSG'] = df.apply(get_consensus_label, axis=1)
    return df
```
